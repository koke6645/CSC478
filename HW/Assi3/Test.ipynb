#### 1b. Perform standard linear regression on data using the implementation for Ch. 8 of MLA.

```python
x = np.array([np.concatenate((v,[1])) for v in vs_minmax])
y = vs_target
print "x:\n", x, "\n\ny:\n", y
```

```python
def standRegres(xArr,yArr):
    xMat = mat(xArr); yMat = mat(yArr).T
    xTx = xMat.T*xMat
    if linalg.det(xTx) == 0.0:
        print "This matrix is singular, cannot do inverse"
        return
    ws = xTx.I * (xMat.T*yMat)
    return ws
```

```python
w = standRegres(x,y)
```

```python
print w[:10]
```

```python
yHat = mat(x)*w
p = yHat.A1
p
```

```python
err = p-y
total_error = np.dot(err,err)
rmse_train = np.sqrt(total_error/len(p))

kf = KFold(len(x), n_folds=10)
xval_err = 0
for train,test in kf:
    standRegres(x[train],y[train])
    yHat_cross = mat(x[test])*w
    pn = yHat_cross.A1
    e = pn-y[test]
    xval_err += np.dot(e,e)
rmse_10cv = np.sqrt(xval_err/len(x))
    
method_name = 'Standard Regression'
print ('Method: %s' %method_name)
print ('RMSE on training: %.4f' %rmse_train)
print ('RMSE on 10-fold: %.4f' %rmse_10cv)

```

```python
pl.plot(p, y, 'ro')
pl.plot([-0.2,1.2],[-0.2,1.2],'g-')
pl.xlabel('predicted')
pl.ylabel('real')
pl.title('Standard Regression')
pl.show()
```

#### 2a. Create your own distance function that, instead of using Euclidean distance, uses Cosine similarity.

```python
def distCosine(vecA, vecB):
    A_norm = np.array([linalg.norm(vecA[i]) for i in range(len(vecA))])
    B_norm = linalg.norm(vecB)
    sims = dot(vecA,vecB)/(A_norm * B_norm)
    return 1 - sims
```

#### 2d. Perform Kmeans clustering on the data.

```python
def randCent(dataSet, k):
    n = shape(dataSet)[1]
    centroids = mat(zeros((k,n)))
    for j in range(n):
        minJ = min(dataSet[:,j])
        rangeJ = float(max(dataSet[:,j]) - minJ)
        centroids[:,j] = minJ + rangeJ * random.rand(k,1)
    return centroids
```

```python
def kMeans(dataSet, k, distMeas=distCosine, createCent=randCent):
    m = shape(dataSet)[0]
    clusterAssment = mat(zeros((m,2)))
    centroids = createCent(dataSet, k)
    clusterChanged = True
    while clusterChanged:
        clusterChanged = False
        for i in range(m):
            minDist = inf; minIndex = -1
            for j in range(k):
                distJI = distMeas(centroids[j,:],dataSet[i,:])
                if distJI < minDist:
                    minDist = distJI; minIndex = j
            if clusterAssment[i,0] != minIndex: clusterChanged = True
            clusterAssment[i,:] = minIndex,minDist**2
        #print centroids 
        for cent in range(k):
            ptsInClust = dataSet[nonzero(clusterAssment[:,0].A==cent)[0]]
            centroids[cent,:] = mean(ptsInClust, axis=0)
    return centroids, clusterAssment
```
