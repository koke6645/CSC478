```python
from sklearn.cross_validation import KFold
from sklearn.feature_extraction import DictVectorizer
from sklearn.linear_model import Ridge
from sklearn import preprocessing
from numpy import *
import numpy as np
import pylab as pl
%pylab inline
```

```python
vs = np.genfromtxt("communities.csv", delimiter=",", names=True, usecols=range(2,100), dtype=float, filling_values="0")
#vs
```

```python
vs_target = vs['ViolentCrimesPerPop']
print vs_target
```

```python
vs = np.genfromtxt("communities.csv", delimiter=",", names=True, usecols=range(2,99), dtype=float, filling_values="0")
```

```python
vs_names = vs.dtype.names
#print vs_names
```

```python
vs_dict = [dict(zip(vs_names, rec)) for rec in vs]
vs_vec = DictVectorizer()
vs_matrix = vs_vec.fit_transform(vs_dict).toarray()
```

```python
minmax_scaler = preprocessing.MinMaxScaler()
vs_minmax = minmax_scaler.fit_transform(vs_matrix)
#vs_minmax
```

```python
print "Mean:\n", np.mean(vs_minmax, axis=0)
print "\nStandard deviation:\n", np.std(vs_minmax, axis=0)
```

```python
x = np.array([np.concatenate((v,[1])) for v in vs_minmax])
y = vs_target
print "x:\n", x, "\n\ny:\n", y
```

```python
def standRegres(xArr,yArr):
    xMat = mat(xArr); yMat = mat(yArr).T
    xTx = xMat.T*xMat
    if linalg.det(xTx) == 0.0:
        print "This matrix is singular, cannot do inverse"
        return
    ws = xTx.I * (xMat.T*yMat)
    return ws
```

```python
w = standRegres(x,y)
```

```python
print w[:10]
```

```python
yHat = mat(x)*w
p = yHat.A1
p
```

```python
err = p-y
total_error = np.dot(err,err)
rmse_train = np.sqrt(total_error/len(p))
print "RMSE train:", rmse_train
```

```python
pl.plot(p, y, 'ro')
pl.plot([-0.2,1.2],[-0.2,1.2],'g-')
pl.xlabel('predicted')
pl.ylabel('real')
pl.show()
```

```python
ridge = Ridge(fit_intercept=True, alpha=0.1)
ridge.fit(x,y)
```

```python
p = ridge.predict(x)
err = p-y
total_error = np.dot(err,err)
rmse_train = np.sqrt(total_error/len(p))
kf = KFold(len(x), n_folds=10)
xval_err = 0

for train,test in kf:
    ridge.fit(x[train],y[train])
    pc = ridge.predict(x[test])
    e = pc-y[test]
    xval_err += np.dot(e,e)
rmse_10cv = np.sqrt(xval_err/len(x))
    
method_name = 'Ridge Regression'
print ('Method: %s' %method_name)
print ('RMSE on training: %.4f' %rmse_train)
print ('RMSE on 10-fold: %.4f' %rmse_10cv)
```

```python
pl.plot(p, y, 'ro')
pl.plot([-0.2,1.2],[-0.2,1.2],'g-')
pl.xlabel('predicted')
pl.ylabel('real')
pl.show()
```
