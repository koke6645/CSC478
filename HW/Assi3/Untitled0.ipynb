```python
from sklearn import preprocessing
from sklearn.cluster import KMeans 
from numpy import *
import numpy as np
import pylab as pl
```

```python
TD = np.genfromtxt('matrix.txt',delimiter=',',dtype=float)
DT = TD.T
```

```python
labels = np.genfromtxt('classes.txt',delimiter=' ',skip_header=1,usecols=(1),dtype=int)
```

```python
DF = np.array([(DT!=0).sum(0)])
NDocs = len(DT[:,0])
NMatrix = np.ones(np.shape(DT), dtype=float)*NDocs
IDF = log2(divide(NMatrix, DF))
DT_tfidf = DT * IDF
print DT_tfidf[0,:]
```

```python
kmeans = KMeans(n_clusters=5, max_iter=500, verbose=1) # initialization
```

```python
kmeans.fit(DT_tfidf)
```

```python
clusters = kmeans.predict(DT_tfidf)
```

```python
np.set_printoptions(precision=2,suppress=True,threshold='nan')
print clusters
```
