```python
from numpy import genfromtxt
import numpy as np
```

```python
vs = np.genfromtxt("bank_data.csv", delimiter=",", names=True, dtype=("|S7", int, float, int, "|S6", "|S10", "|S3", "|S3", "|S3", "|S3", "|S3", "|S3"))
vs
```

```python
vs_records = vs[['age','income','children','gender','region','married','car','savings_acct','current_acct','mortgage','pep']]
vs_records
```

```python
vs_names = vs_records.dtype.names
vs_names
```

```python
vs_target = vs['pep']
vs_target
```

```python
vs_dict = [dict(zip(vs_names, record)) for record in vs_records]
```

```python
vs_dict[:5]
```

```python
vs_dict[0]['age']
```

```python
from sklearn.feature_extraction import DictVectorizer
```

```python
vs_vec = DictVectorizer()
vs_matrix = vs_vec.fit_transform(vs_dict).toarray()
```

```python
vs_vec.get_feature_names()
```

```python
print vs_matrix[:5]
```

```python
tpercent = 0.8
tsize = tpercent * len(vs_matrix)
vs_train = vs_matrix[:tsize,:]
vs_test = vs_matrix[tsize:,:]
target_train = vs_target[:tsize]
target_test = vs_target[tsize:]
```

#### b. Run scikit-learn's KNN classifier on the test set.

```python
from sklearn import preprocessing
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
```

```python
vs_train_norm = min_max_scaler.fit_transform(vs_train)
vs_test_norm = min_max_scaler.fit_transform(vs_test)
```

```python
vs_train_norm[:5]
```

```python
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
```

```python
clf = KNeighborsClassifier(n_neighbors=20, weights='distance', algorithm='auto')
clf.fit(vs_train_norm, target_train)
predicted = clf.predict(vs_test_norm)
print metrics.confusion_matrix(target_test, predicted)
print metrics.classification_report(target_test, predicted)
```
